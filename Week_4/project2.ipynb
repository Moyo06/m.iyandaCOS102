```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df.drop(columns=['Cabin', 'Ticket', 'Name'], inplace=True, errors='ignore')
```

```python
df.head()
```

```python
df.tail()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Sex', y='Survived', data=df, palette='magma', hue='Sex', legend=False)
plt.title('Survival Rate: Men vs Women')
plt.show()
```

INSIGHT: It was women and children first...the graoh shows a large number of females surviving. Meaning that, only a few males survived.

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Pclass', y='Survived', data=df, palette='viridis', hue='Pclass', legend=False)
plt.title('Survival Rate by Ticket Class')
plt.show()
```

INSIGHT: Money definitely played a part in who lived. If you had a 1st Class ticket (thexpensive ones), you were much more likely to get on a lifeboat than the people in 3rd Class.

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Age'], color='skyblue')
plt.title('How Old Were the Passengers?')
plt.xlabel('Age')
plt.ylabel('Number of People')
plt.show()
```

INSIGHT: The graph shows a major 'spike' right around age 30. This tells us that the Titanic was packed with young adults and people in their prime. While there were some kids and some older people, the huge 'mountain' in the middle shows that 30-year-olds were the biggest group on the journey.
