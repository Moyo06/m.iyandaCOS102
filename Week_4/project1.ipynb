```python
plt.figure(figsize= (8, 5))
plt.title('Locality Distribution')
sns.scatterplot(data=df, x='number_of_reviews', y='price', hue='neighbourhood_group')
plt.show()
```

```python
INSIGHT: The cheap spots get all the love. You can see that the places with the most reviews are almost always the lower-priced ones. The really expensive places hardly have any reviews at all. Basically, if you want a lot of customers in New York City, you have to keep your price low.
```
